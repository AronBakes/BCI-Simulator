```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
eeg_data = np.load("data/processed/eeg_processed.npz")
X_lstm, y = eeg_data['windows'], eeg_data['labels']

image_data = np.load("data/raw/brain_images.npz")
X_image = image_data['images']

text_data = np.load("data/raw/text_data.npz")
X_text, texts = text_data['features'], text_data['texts']

# Visualize one sample
sample_idx = 0

# EEG
plt.figure(figsize=(12, 6))
for ch in range(X_lstm.shape[2]):
    plt.plot(X_lstm[sample_idx, :, ch], label=f'Channel {ch+1}')
plt.title(f'EEG Sample {sample_idx} (Label: {y[sample_idx]})')
plt.xlabel('Time Points')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Image
plt.figure(figsize=(6, 6))
plt.imshow(X_image[sample_idx], cmap='gray')
plt.title(f'Brain Image Sample {sample_idx}')
plt.axis('off')
plt.show()

# Text
print(f"Text Sample {sample_idx}: {texts[sample_idx]}")
print(f"Text Features Shape: {X_text[sample_idx].shape}")

# Visualize multiple samples
n_samples = min(5, len(X_lstm))
fig, axes = plt.subplots(n_samples, 2, figsize=(12, 6*n_samples))
for i in range(n_samples):
    axes[i, 0].plot(X_lstm[i, :, 0])
    axes[i, 0].set_title(f'EEG Sample {i} (Label: {y[i]})')
    axes[i, 0].set_xlabel('Time Points')
    axes[i, 0].grid(True)
    axes[i, 1].imshow(X_image[i], cmap='gray')
    axes[i, 1].set_title(f'Image Sample {i}')
    axes[i, 1].axis('off')
plt.tight_layout()
plt.show()
```
